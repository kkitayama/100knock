92„Åß‰Ωú„Å£„Åü„Éá„Éº„Çø„ÇíÁî®„ÅÑÔºåÂêÑ„É¢„Éá„É´„ÅÆ„Ç¢„Éä„É≠„Ç∏„Éº„Çø„Çπ„ÇØ„ÅÆÊ≠£Ëß£Áéá„ÇíÊ±Ç„ÇÅ„ÇàÔºé

```python
def calc_accuracy(result_text):
    ALL = 0
    T = 0
    with open(result_text, 'r') as f:
        for line in f:
            ALL += 1
            words = line.strip().split()
            if words[3] == words[4]:
                T += 1
    return T / ALL
```

### „Çπ„Éî„Ç¢„Éû„É≥Áõ∏Èñ¢‰øÇÊï∞:
$$\rho =1-{{\frac  {6\sum D^{2}}{N^{3}-N}}}$$
D = ÂØæÂøú„Åô„ÇãX „Å®Y „ÅÆÂÄ§„ÅÆÈ†Ü‰Ωç„ÅÆÂ∑Æ

N = ÂÄ§„ÅÆ„Éö„Ç¢„ÅÆÊï∞

https://ja.wikipedia.org/wiki/%E3%82%B9%E3%83%94%E3%82%A2%E3%83%9E%E3%83%B3%E3%81%AE%E9%A0%86%E4%BD%8D%E7%9B%B8%E9%96%A2%E4%BF%82%E6%95%B0

### scipy.stats.spearmanr
> nan_policy : {‚Äòpropagate‚Äô, ‚Äòraise‚Äô, ‚Äòomit‚Äô}, optional
Defines how to handle when input contains nan. ‚Äòpropagate‚Äô returns nan, ‚Äòraise‚Äô throws an error, ‚Äòomit‚Äô performs the calculations ignoring nan values. Default is ‚Äòpropagate‚Äô.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html

```python
from scipy.stats import spearmanr

def get_spearman_correlation(result_text):
    a = []
    b = []
    with open(result_text, 'r') as f:
        for line in f:
            words = line.strip().split()
            a.append(float(words[2]))
            b.append(float(words[3]))
        return spearmanr(a, b, nan_policy='omit')[0]
```

## 98. WardÊ≥ï„Å´„Çà„Çã„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞

96„ÅÆÂçòË™û„Éô„ÇØ„Éà„É´„Å´ÂØæ„Åó„Å¶ÔºåWardÊ≥ï„Å´„Çà„ÇãÈöéÂ±§Âûã„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞„ÇíÂÆüË°å„Åõ„ÇàÔºé„Åï„Çâ„Å´Ôºå„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞ÁµêÊûú„Çí„Éá„É≥„Éâ„É≠„Ç∞„É©„É†„Å®„Åó„Å¶ÂèØË¶ñÂåñ„Åõ„ÇàÔºé

### Python „ÅßÈöéÂ±§Âûã„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞
https://gomiba.co.in/blog/archives/705

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# ÈöéÂ±§Âûã„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞„ÅÆÂÆüÊñΩ
# „Ç¶„Ç©„Éº„ÉâÊ≥ï x „É¶„Éº„ÇØ„É™„ÉÉ„ÉâË∑ùÈõ¢
linkage_result = linkage(countries_array, method='ward', metric='euclidean')

# „ÇØ„É©„Çπ„ÇøÂàÜ„Åë„Åô„Çã„Åó„Åç„ÅÑÂÄ§„ÇíÊ±∫„ÇÅ„Çã
threshold = 0.4 * np.max(linkage_result[:, 2])

# ÈöéÂ±§Âûã„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞„ÅÆÂèØË¶ñÂåñ
plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(linkage_result, labels=df_country_vec.columns, color_threshold=threshold)
plt.show()

# „ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞ÁµêÊûú„ÅÆÂÄ§„ÇíÂèñÂæó
clustered = fcluster(linkage_result, threshold, criterion='distance')

# „ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞ÁµêÊûú„ÇíÊØîËºÉ
print(clustered)
```
