## 92. アナロジーデータへの適用

91で作成した評価データの各事例に対して，vec(2列目の単語) - vec(1列目の単語) + vec(3列目の単語)を計算し，そのベクトルと類似度が最も高い単語と，その類似度を求めよ．求めた単語と類似度は，各事例の末尾に追記せよ．このプログラムを85で作成した単語ベクトル，90で作成した単語ベクトルに対して適用せよ．

```python
model = word2vec.Word2Vec.load("./work/word2vec.model")
```

```python
dir(model.wv)
```

```python
'Anarchism' in model.wv.vocab
```

```python
with open('./work/q092by90', 'w') as fw:
    with open('./work/q091.txt', 'r') as fr:
        for line in fr:
            words = line.strip().split()
            if words[0] in model.wv.vocab and words[1] in model.wv.vocab and words[2] in model.wv.vocab:   # vocabの中に単語が含まれている時
                most_similar_word = model.wv.most_similar(positive=[words[1], words[2]], negative=[words[0]])[0]
                fw.write(line.strip() + ' ' + most_similar_word[0] + ' ' + str(most_similar_word[1]) + '\n')
```

## 93. アナロジータスクの正解率の計算

92で作ったデータを用い，各モデルのアナロジータスクの正解率を求めよ．

```python
def calc_accuracy(result_text):
    ALL = 0
    T = 0
    with open(result_text, 'r') as f:
        for line in f:
            ALL += 1
            words = line.strip().split()
            if words[3] == words[4]:
                T += 1
    return T / ALL
```

## 95. WordSimilarity-353での評価

94で作ったデータを用い，各モデルが出力する類似度のランキングと，人間の類似度判定のランキングの間のスピアマン相関係数を計算せよ．

### スピアマン相関係数:
$$\rho =1-{{\frac  {6\sum D^{2}}{N^{3}-N}}}$$
D = 対応するX とY の値の順位の差

N = 値のペアの数

https://ja.wikipedia.org/wiki/%E3%82%B9%E3%83%94%E3%82%A2%E3%83%9E%E3%83%B3%E3%81%AE%E9%A0%86%E4%BD%8D%E7%9B%B8%E9%96%A2%E4%BF%82%E6%95%B0

### scipy.stats.spearmanr
> nan_policy : {‘propagate’, ‘raise’, ‘omit’}, optional
Defines how to handle when input contains nan. ‘propagate’ returns nan, ‘raise’ throws an error, ‘omit’ performs the calculations ignoring nan values. Default is ‘propagate’.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html

```python
from scipy.stats import spearmanr

def get_spearman_correlation(result_text):
    a = []
    b = []
    with open(result_text, 'r') as f:
        for line in f:
            words = line.strip().split()
            a.append(float(words[2]))
            b.append(float(words[3]))
        return spearmanr(a, b, nan_policy='omit')[0]
```

## 98. Ward法によるクラスタリング

96の単語ベクトルに対して，Ward法による階層型クラスタリングを実行せよ．さらに，クラスタリング結果をデンドログラムとして可視化せよ．

### Python で階層型クラスタリング
https://gomiba.co.in/blog/archives/705

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# 階層型クラスタリングの実施
# ウォード法 x ユークリッド距離
linkage_result = linkage(countries_array, method='ward', metric='euclidean')

# クラスタ分けするしきい値を決める
threshold = 0.4 * np.max(linkage_result[:, 2])

# 階層型クラスタリングの可視化
plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(linkage_result, labels=df_country_vec.columns, color_threshold=threshold)
plt.show()

# クラスタリング結果の値を取得
clustered = fcluster(linkage_result, threshold, criterion='distance')

# クラスタリング結果を比較
print(clustered)
```
